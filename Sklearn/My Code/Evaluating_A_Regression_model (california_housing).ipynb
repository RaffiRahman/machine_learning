# **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing.data.shape, housing.target.shape)
print(housing.feature_names[0:6])
```

```python
# creating dataframe

housing_df = pd.DataFrame(housing['data'], columns = housing['feature_names'])
housing_df
```

```python
housing_df.columns
```

```python
# adding the Target column

housing_df['target'] = housing['target']
housing_df
```

# **Identifying Estimator**

```python
# importing train_test_split
from sklearn.model_selection import train_test_split
```

## **Trying Ridge**

```python
# import Ridge
from sklearn.linear_model import Ridge
```

```python
# random seed
np.random.seed(42)
```

```python
# creating X and y
X = housing_df.drop('target', axis = 1)
y = housing_df['target']
```

```python
# spliting into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# instabtiate and fit the model (on the training set)
model = Ridge()
model.fit(X_train, y_train)
```

```python
# check the score of the model (on the test sets)
model.score(X_test, y_test)
```

## **Trying RandomForestRegressor**

```python
# import RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor

# setup random seed
np.random.seed(42)
```

```python
# creating X and y
X = housing_df.drop('target', axis = 1)
y = housing_df['target']

# spliting into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# instabtiate and fit the model (on the training set)
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
# check the score of the model (on the test sets)
model.score(X_test, y_test)
```

# **R2 Score**

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
# r2 score using the same value as y_pred
# Fill an array with y_test mean
y_test_mean = np.full(len(y_test), y_test.mean())

y_test_mean[:10]

r2_score(y_true = y_test, y_pred = y_test_mean)
```

```python
# y_pred is the actual prediction
y_pred = model.predict(X_test)
y_pred
```

```python
# r2 score using y_pred
r2_score(y_true = y_test, y_pred = y_pred)
```

# **Mean Absolute Error (MAE)**

MAE is the average of the absolute difference between predictions and actual values.
it gives you an idea of how wrong your predictions are.

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
df = pd.DataFrame(data = {"actual values": y_test, "predicted values": y_pred})
df["difference"] = df["predicted values"] - df["actual values"]
df.head(10)
```

```python
df["difference"].mean()
```

```python
df["difference"].abs().mean()
```

# **Mean Squared Error (MSE)**

MSE is the mean of the square of the errors between actual and predicted values.

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
df["squared_difference"] = df["difference"]**2
df.head()
```

```python
#calculate mse by hand
squared = np.square(df["difference"])
squared.mean()
```
