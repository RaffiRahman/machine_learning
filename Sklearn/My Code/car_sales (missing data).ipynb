# **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Reading csv**

```python
car_sales_missing = pd.read_csv('car-sales-extended-missing-data.csv')
car_sales_missing.isnull().sum()
```

# **Dropping 'Price' with null values**

```python
car_sales_missing.dropna(subset= ['Price'], inplace= True)
car_sales_missing.isnull().sum()
```

# **Splitting X and y**

```python
X = car_sales_missing.drop('Price', axis=1)

y = car_sales_missing['Price']
```

# **Filling the Missing Values**

```python
# iporting SimpleImputer and ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
X.head()
```

```python
# filling categorical values with 'missing' and numeric values with mean
cat_imputer = SimpleImputer(strategy= 'constant', fill_value= 'missing')
door_imputer = SimpleImputer(strategy= 'constant', fill_value= 4)
num_imputer = SimpleImputer(strategy= 'mean')

# define column
cat_features = ['Make', 'Colour']
door_features = ['Doors']
num_features = ['Odometer (KM)']

# Create an imputer (something that fill the missing values)

imputer = ColumnTransformer([('cat_imputer', cat_imputer, cat_features),
                             ('door_imputer', door_imputer, door_features),
                             ('num_imputer', num_imputer, num_features)])

# Transform the data
filled_X = imputer.fit_transform(X)
filled_X
```

```python
car_sales_filled = pd.DataFrame(filled_X,
                                columns= ['Make','Colour','Odometer (KM)','Doors'])
car_sales_filled
```
