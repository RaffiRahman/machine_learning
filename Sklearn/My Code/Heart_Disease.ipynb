# **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Reading csv file**

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

# **Creating X, Y**

```python
#Create X (feature matrix)
X = heart_disease.drop('target', axis=1)

#Create Y (labels)
y = heart_disease['target']

#random seed
np.random.seed(42)
```

# **Choose the right model and hyperparameters**

```python
# importing cross_val_score
from sklearn.model_selection import cross_val_score
# importing RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
```

# **Fit the model to the training data and make the prediction**

```python
# importing train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
clf.fit(X_train, y_train);
```

```python
X_train
```

```python
# make predistion
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

# **Evaluate the Model on the Training Data and Test Data**

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
# importing classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

# **Improve a Model**

```python
# try different amount of n_estimators

np.random.seed(42)
for i in range(10, 100, 10):
  print(f"Trying model with {i} estimators...")
  clf = RandomForestClassifier(n_estimators = i).fit(X_train, y_train)
  print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
  print(" ")
```

# **Save a Model and Load It**

```python
# importing pickle
import pickle
```

```python
pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
loading_model = pickle.load(open("random_forest_model_1.pkl", "rb"))
loading_model.score(X_test, y_test)
```
