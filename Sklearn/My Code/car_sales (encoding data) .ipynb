# **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Reading csv file**

```python
car_sales = pd.read_csv('car-sales-extended.csv')
car_sales
```

# **Creating X,y**

```python
car_sales.head()
```

```python
#Create X (feature matrix)
X = car_sales.drop('Price', axis=1)

#Create Y (labels)
y = car_sales['Price']
```

# **Encoding data**

```python
# importing OneHotEncoder and ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Encoding the string values
categorical_features = ['Make', 'Colour', 'Doors']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)],
                                remainder = 'passthrough')

transformer_X = transformer.fit_transform(X)
transformer_X
```

```python
X.head()
```

```python
pd.DataFrame(transformer_X)
```

# **Tranning the Model**

```python
# importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(transformer_X, y, test_size = 0.2)
```

# **Build Model**

```python
# importing RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
